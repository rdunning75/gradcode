```python
def cropTiff(path, save_path, window, dirs):
    for item in dirs:
        fullpath = os.path.join(path,item)
        if os.path.isfile(fullpath) and fullpath.endswith('.tif'):
            ds = gdal.Open(fullpath)
#             print(type(ds))
            f, e = os.path.splitext(item) 
            print("cropping " + item)
            f = f+"_cropped.tif"
#             print(f)
            save_path_final = os.path.join(save_path,f)
#             print(save_path_final)
            ds = gdal.Translate(save_path_final, ds, projWin = window)
            ds = None
    print("Done Cropping!")
            
```

```python
path1NDVITop =r"D:\CS_Course_Material\Graduate_Program\workspace\DATA\GEOTIFF\NDVI\Bottom"
save_pathNDVITop = r"D:\CS_Course_Material\Graduate_Program\workspace\DATA\GEOTIFF\NDVI\CroppedBottom"
windowNDVITop = [-123,38.3, -117.94,35.65]
dirsNDVITop = os.listdir(path1NDVITop)

cropTiff(path1NDVITop,save_pathNDVITop,windowNDVITop,dirsNDVITop)
```

```python
pathFireMap = r"D:\CS_Course_Material\Graduate_Program\workspace\DATA\GEOTIFF\FireMask"
save_pathFireMap = r"D:\CS_Course_Material\Graduate_Program\workspace\DATA\GEOTIFF\FireMask\CroppedFireMask"
windowFire = [-123,38.3, -117.94,35.65]

dirsFireMap= os.listdir(pathFireMap)
cropTiff(pathFireMap,save_pathFireMap,windowFire,dirsFireMap)
```
